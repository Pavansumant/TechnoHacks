```python
class_counts = dataset["Class"].value_counts()

# Data for the pie chart
counts = class_counts.values
labels = class_counts.index
colors = ['skyblue', 'salmon']
explode = (0, 0.1)  # Explode the second slice (Fraudulent)

# Create the pie chart
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(counts, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.title("Distribution of Credit Card Transaction Classes")
plt.show()
```

```python
X = dataset.iloc[: , :-1].values
y = dataset.iloc[: , -1].values
```

```python
#convert imbalanced data to balanced data
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
x_res , y_res = ros.fit_resample(X,y)
```

```python
X.shape
```

```python
x_res.shape,y_res.shape
```

```python
from collections import Counter
print(Counter(y))
print(Counter(y_res))
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_res , y_res , test_size=0.3 , random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators =10 , random_state=0)
classifier.fit(x_train , y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
n_errors = (y_pred != y_test).sum()
```

```python
n_errors
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(y_test , y_pred)
sns.heatmap(cm , annot=True)
print(accuracy_score(y_test , y_pred))
```

```python
from sklearn.metrics import precision_score
precision_score(y_test , y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test , y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred))
```
