```python
d={
    'setosa':0,
    'versicolor':1,
    'virginica':2
}
```

Machine learning algorithm takes numerical features.so converted species to numerical column 

```python
df["species"]=df["species"].map(d)
df
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['species'], axis='columns')
y = df.species

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
cl = LogisticRegression(random_state = 0)
cl.fit(x_train, y_train)
```

```python
y_pred=cl.predict(x_test)
y_pred
```

```python
cl.score(x_train, y_train)
```

```python
cl.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot=True, annot_kws={"size": 30}, fmt='d',cmap="Blues")
```
